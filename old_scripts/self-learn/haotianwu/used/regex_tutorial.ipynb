## 1. 导入 re 模块

Python 内置 `re` 模块，无需安装。

````python
import re

# 最简单的匹配：在字符串中查找 "hello"
text = "hello world, hello python"
result = re.search(r"hello", text)
print(f"找到: '{result.group()}', 位置: {result.span()}")
````

## 2. 常用元字符

| 元字符 | 含义 | 示例 |
|--------|------|------|
| `.` | 匹配任意单个字符（除换行） | `a.c` → `abc`, `a1c` |
| `^` | 匹配字符串开头 | `^hello` |
| `$` | 匹配字符串结尾 | `world$` |
| `\d` | 匹配数字 `[0-9]` | `\d+` → `123` |
| `\w` | 匹配字母/数字/下划线 | `\w+` → `hello_1` |
| `\s` | 匹配空白字符 | `\s+` |
| `\b` | 单词边界 | `\bword\b` |
| `[]` | 字符集合 | `[aeiou]` 匹配元音 |
| `[^]` | 排除字符集合 | `[^0-9]` 匹配非数字 |

````python
# ─── 2.1 点号 . ：匹配任意字符 ──────────────────────────────
text = "cat cot cut c3t"
matches = re.findall(r"c.t", text)
print(f". 匹配: {matches}")  # ['cat', 'cot', 'cut', 'c3t']
````

````python
# ─── 2.2 ^ 和 $ ：开头和结尾 ──────────────────────────────
text = "hello world"
match=re.search(r"^hello", text)
if match:
    print(match.group(0))  # 匹配成功
else:
    print("未匹配到开头的 'hello'")
print(re.search(r"^world", text))  # None（world不在开头）
print(re.search(r"world$", text))  # 匹配成功
````

````python
# ─── 2.3 \d \w \s ：字符类快捷方式 ───────────────────────
text = "Gene AT1G01010 has 3 exons"

digits = re.findall(r"\d+", text)
print(f"数字: {digits}")  # ['1', '01010', '3']

words = re.findall(r"\w+", text)
print(f"单词: {words}")  # ['Gene', 'AT1G01010', 'has', '3', 'exons']

spaces = re.findall(r"\s+", text)
print(f"空白: {spaces}")  # [' ', ' ', ' ', ' ']
````

````python
# ─── 2.4 [] 字符集合 ─────────────────────────────────────
text = "apple banana+a1 cherry date"

# 匹配以 a/b/c 开头的单词
matches = re.findall(r"\b[abc]\w+", text)
print(f"以 a/b/c 开头的单词: {matches}")  # ['apple', 'banana', 'cherry']
````

## 3. 量词（重复匹配）

| 量词 | 含义 | 示例 |
|------|------|------|
| `*` | 0次或多次 | `ab*c` → `ac`, `abc`, `abbc` |
| `+` | 1次或多次 | `ab+c` → `abc`, `abbc`（不匹配 `ac`）|
| `?` | 0次或1次 | `colou?r` → `color`, `colour` |
| `{n}` | 恰好n次 | `\d{3}` → `123` |
| `{n,m}` | n到m次 | `\d{2,4}` → `12`, `123`, `1234` |
| `{n,}` | 至少n次 | `\d{2,}` → `12`, `123`, `12345` |

````python
# ─── 3.1 基本量词 ────────────────────────────────────────
text = "ac abc abbc abbbc"

print(f"ab*c  (0+次b): {re.findall(r'ab*c', text)}")
print(f"ab+c  (1+次b): {re.findall(r'ab+c', text)}")
print(f"ab?c  (0或1次b): {re.findall(r'ab?c', text)}")
````

````python
# ─── 3.2 精确量词 {n} {n,m} ──────────────────────────────
text = "Tel: 010-12345678, Zip: 100001, Code: 42"

print(f"恰好3位数字: {re.findall(r'\d{3}', text)}")
print(f"5-8位数字:   {re.findall(r'\d{5,8}', text)}")
````

````python
text = "hello world"
a=re.findall(r'^hell ',text)
print(a)  # ['hell']，匹配开头的 'hell'
````

````python
# ─── 3.3 贪婪 vs 非贪婪（懒惰）─────────────────────────
text = "<div>hello</div><div>world</div>"

# 贪婪模式（默认）: 尽可能多匹配
greedy = re.findall(r"<div>.*</div>", text)
print(f"贪婪: {greedy}")

# 非贪婪模式（加 ?）: 尽可能少匹配
lazy = re.findall(r"<div>.*?</div>", text)
print(f"非贪婪: {lazy}")
````

## 4. 分组与捕获

用 `()` 创建分组，可以：
- 提取匹配的子串
- 对分组使用量词
- 在替换中引用分组

````python
# ─── 4.1 基本分组 ────────────────────────────────────────
text = "2026-03-15 10:30:00"

# 提取日期的年、月、日
match = re.search(r"(\d{4})-(\d{2})-(\d{2})", text)
if match:
    print(f"完整匹配: {match.group(0)}")
    print(f"年: {match.group(1)}")
    print(f"月: {match.group(2)}")
    print(f"日: {match.group(3)}")
    print(f"所有分组: {match.groups()}")
````

````python
# ─── 4.2 命名分组 (?P<name>...) ─────────────────────────
text = "Gene: AT1G01010, Expression: 45.3 FPKM"

pattern = r"Gene: (?P<gene>\w+), Expression: (?P<value>[\d.]+) (?P<unit>\w+)"
match = re.search(pattern, text)
if match:
    print(f"基因: {match.group('gene')}")
    print(f"表达量: {match.group('value')} {match.group('unit')}")
    print(f"字典形式: {match.groupdict()}")
````

````python
# ─── 4.3 findall + 分组 ─────────────────────────────────
text = "scores: 85-90, 70-75, 92-98"

# findall 有分组时，返回分组内容（元组列表）
pairs = re.findall(r"(\d+)-(\d+)", text)
print(f"配对: {pairs}")

for low, high in pairs:
    print(f"  范围: {low} ~ {high}, 中间值: {(int(low)+int(high))/2}")
````

## 5. re 模块核心函数

| 函数 | 作用 | 返回值 |
|------|------|--------|
| `re.search(pattern, text)` | 找第一个匹配 | Match 对象 或 None |
| `re.match(pattern, text)` | 从**开头**匹配 | Match 对象 或 None |
| `re.findall(pattern, text)` | 找所有匹配 | 列表 |
| `re.finditer(pattern, text)` | 找所有匹配（迭代器） | Match 对象迭代器 |
| `re.sub(pattern, repl, text)` | 替换匹配 | 替换后的字符串 |
| `re.split(pattern, text)` | 按模式分割 | 列表 |
| `re.compile(pattern)` | 预编译模式 | Pattern 对象 |

````python
# ─── 5.1 re.search vs re.match ──────────────────────────
text = "hello world"

# search: 在任意位置查找
print(f"search 'world': {re.search(r'world', text)}")

# match: 只在开头匹配
print(f"match 'world':  {re.match(r'world', text)}")
print(f"match 'hello':  {re.match(r'hello', text)}")
````

````python
# ─── 5.2 re.findall ─────────────────────────────────────
text = "apple 3, banana 5, cherry 12"

# 无分组: 返回匹配的字符串列表
all_numbers = re.findall(r"\d+", text)
print(f"所有数字: {all_numbers}")

# 有分组: 返回分组内容
fruit_count = re.findall(r"(\w+) (\d+)", text)
print(f"水果-数量: {fruit_count}")
````

````python
# ─── 5.3 re.finditer（推荐处理大量匹配）──────────────────
text = "Gene AT1G01010 on chr1, Gene AT3G55120 on chr3"

for match in re.finditer(r"AT\dG\d{5}", text):
    print(f"  基因 {match.group()} 位于位置 {match.span()}")
````

````python
# ─── 5.4 re.sub（替换）───────────────────────────────────
text = "Phone: 138-1234-5678, Tel: 010-8765-4321"

# 隐藏手机号中间4位
masked = re.sub(r"(\d{3})-\d{4}-(\d{4})", r"\1-****-\2", text)
print(f"脱敏后: {masked}")
````

````python
# ─── 5.5 re.split（按模式分割）───────────────────────────
text = "apple, banana;  cherry|date   elderberry"

# 按多种分隔符分割
items = re.split(r"[,;|]?\s+", text)
print(f"分割结果: {items}")
````

````python
# ─── 5.6 re.compile（预编译，提高效率）────────────────────
# 当同一个模式需要反复使用时，预编译更高效
gene_pattern = re.compile(r"AT\dG\d{5}")

texts = [
    "AT1G01010 encodes a NAC domain protein",
    "The gene AT5G44790 is involved in flowering",
    "No gene ID mentioned here",
]

for t in texts:
    match = gene_pattern.search(t)
    if match:
        print(f"  ✅ 找到基因: {match.group()}")
    else:
        print(f"  ❌ 未找到基因")
````

## 6. 实战案例

结合生物信息学和日常开发的真实场景。

### 6.1 提取 Markdown 中的基因名

````python
md_text = """
## Results
The gene **AT1G01010** (NAC domain) showed 2.5-fold increase.
Meanwhile, **AT3G55120** and **AT5G44790** were downregulated.
Os01g0100100 is a rice gene not matching Arabidopsis pattern.
"""

# 匹配拟南芥基因 ID 格式: AT + 数字 + G + 5位数字
genes = re.findall(r"AT\dG\d{5}", md_text)
print(f"找到的拟南芥基因: {genes}")
````

### 6.2 解析 JSON 中的数值字段

````python
json_text = '{"gene": "AT1G01010", "expression": 45.3, "p_value": 0.001, "fold_change": -2.5}'

# 提取所有浮点数（包括负数）
numbers = re.findall(r"-?\d+\.\d+", json_text)
print(f"所有数值: {numbers}")

# 提取键值对
pairs = re.findall(r'"(\w+)":\s*([^,}]+)', json_text)
for key, val in pairs:
    print(f"  {key}: {val.strip('"')}")
````

### 6.3 清洗文本：去除多余空白和特殊字符

````python
dirty_text = "  Hello   World!\n\n  This  has   extra    spaces.  \t\n"

# 1. 去除首尾空白
clean = dirty_text.strip()

# 2. 多个空白合并为一个空格
clean = re.sub(r"\s+", " ", clean)
print(f"清洗后: '{clean}'")
````

### 6.4 验证邮箱格式

````python
emails = [
    "user@example.com",
    "invalid-email",
    "user.name+tag@domain.co.uk",
    "@no-user.com",
    "test@.com",
]

email_pattern = re.compile(r"^[\w.+-]+@[\w-]+\.[\w.]+$")

for email in emails:
    valid = "✅" if email_pattern.match(email) else "❌"
    print(f"  {valid} {email}")
````

### 6.5 从文件名中提取信息

````python
filenames = [
    "MinerU_markdown_mmc10_2031567061837611008.md",
    "MinerU_markdown_PIIS009286741731499X_2031567088198815744.md",
    "token_usage_extract_20260314_150807.json",
]

# 提取时间戳格式 YYYYMMDD_HHMMSS
for f in filenames:
    ts = re.search(r"(\d{8})_(\d{6})", f)
    if ts:
        date, time = ts.groups()
        print(f"  {f}")
        print(f"    日期: {date[:4]}-{date[4:6]}-{date[6:]}")
        print(f"    时间: {time[:2]}:{time[2:4]}:{time[4:]}")
    else:
        print(f"  {f} — 无时间戳")
````

## 7. 常见陷阱与注意事项

### ⚠️ 陷阱1: 忘记用 raw string
```python
# ❌ 错误: \b 会被 Python 解释为退格符
re.search("\bword\b", text)

# ✅ 正确: 使用 r"" 前缀
re.search(r"\bword\b", text)
```

### ⚠️ 陷阱2: 贪婪匹配导致意外结果
```python
# ❌ .* 会匹配到最后一个 </div>
re.findall(r"<div>.*</div>", html)

# ✅ 用 .*? 非贪婪
re.findall(r"<div>.*?</div>", html)
```

### ⚠️ 陷阱3: findall 有分组时只返回分组内容
```python
# 只返回分组 (\d+)，不返回完整匹配
re.findall(r"price: (\d+)", "price: 100")  # ['100']

# 想要完整匹配，去掉分组或用 (?:...)
re.findall(r"price: \d+", "price: 100")    # ['price: 100']
```

## 8. 正则表达式速查表

```
字符类:    .  \d  \D  \w  \W  \s  \S  [abc]  [^abc]  [a-z]
量词:      *  +  ?  {n}  {n,m}  {n,}
非贪婪:    *?  +?  ??  {n,m}?
锚点:      ^  $  \b  \B
分组:      (...)  (?P<name>...)  (?:...)  \1  \2
或:        a|b
前瞻/后顾: (?=...)  (?!...)  (?<=...)  (?<!...)
标志:      re.IGNORECASE  re.MULTILINE  re.DOTALL
```
